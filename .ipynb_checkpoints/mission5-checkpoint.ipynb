```python
from datetime import datetime,timedelta

import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rc('font', family='nanumgothic')
plt.rc('axes', unicode_minus=False)
```

```python
df_basic= pd.read_excel('./data/project data/기본 데이터/0_유동인구_유동인구기본_2015.xlsx',sheet_name=None,skiprows=[0,1,3])
df_basic
```

```python
df_detail= pd.read_excel('./data/project data/기본 데이터/2_유동인구_속성조사_2015.xlsx',sheet_name=None,skiprows=[0,1,3])
df_detail
```

```python
df_pop,df_popSpot=df_basic.values()
df_walk,df_walkSpot=df_detail.values()
```

```python
df_pop.head()
```

```python
df_walk.head()
```

```python
df_walkEnv=df_walk[['조사지점코드','보행환경']]
df_walkEnv=df_walkEnv.set_index('조사지점코드').join(df_walkSpot.set_index('조사지점코드'))
df_walkEnv=df_walkEnv.reset_index()
df_walkEnv
```

```python
def createEnvScore(sr):
    if sr['보행환경']=='매우불만족':
        return 1
    elif sr['보행환경']=='약간불만족':
        return 2
    elif sr['보행환경']=='보통':
        return 3
    elif sr['보행환경']=='약간만족':
        return 4
    elif sr['보행환경']=='매우만족':
        return 5
    else:
        return -1

df_walkEnv['보행환경점수']=df_walkEnv.apply(createEnvScore,axis=1)
df_walkEnv
```

```python
df_walkEnv=df_walkEnv.groupby(['조사지점명']).sum().sort_values(by='보행환경점수',ascending=True)
df_walkEnv
```

```python
df_popNum=df_pop[['조사지점코드','유동인구수']]
df_popNum
```

```python
df_popNum=df_popNum.set_index('조사지점코드').join(df_popSpot.set_index('조사지점코드'))
df_popNum
```

```python
df_popNum=df_popNum.fillna('unknown')
df_popNum
```

```python
df_popNum=df_popNum.reset_index().groupby('조사지점명').sum()
df_popNum
```

```python
df_popNum=df_popNum.drop(['unknown'])
df_popNum
```

```python
df_result=df_walkEnv.join(df_popNum)
df_result
```

```python
df_result=df_result.sort_values(['보행환경점수', '유동인구수'], ascending=[True,False])
df_result
```

```python
df_result['유동인구수'].mean()
```

```python
df_resultPlot=df_result[df_result['유동인구수'] >= df_result['유동인구수'].mean()][:10]
df_resultPlot
```

```python
df_resultPlot['보행환경점수'].plot(kind = 'barh',stacked=False, fontsize = 13,figsize=(10, 10))
```

```python
df_resultPlot['유동인구수'].plot(kind = 'barh',stacked=False, fontsize = 13,figsize=(10, 10))
```
