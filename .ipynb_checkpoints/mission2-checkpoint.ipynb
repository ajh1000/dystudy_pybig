```python
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_excel('./data/project data/기본 데이터/0_유동인구_유동인구기본_2015.xlsx',sheet_name=None,skiprows=[0,1,3])
df
```

```python
df_pop,df_spot=df.values()
```

```python
df_popInSpot=df_pop[['유동인구수','조사지점코드']]
df_popInSpot
```

```python
df_popInSpot=df_popInSpot.groupby('조사지점코드').agg('sum')
df_popInSpot
```

```python
df_result=df_popInSpot.join(df_spot.set_index('조사지점코드'))
df_result
```

```python
df_result=df_result.reset_index().set_index('조사지점명')
df_result
```
