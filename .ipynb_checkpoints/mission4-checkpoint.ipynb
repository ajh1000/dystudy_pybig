```python
from datetime import datetime,timedelta

import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rc('font', family='nanumgothic')
plt.rc('axes', unicode_minus=False)
```

```python
df= pd.read_excel('./data/project data/기본 데이터/0_유동인구_유동인구기본_2015.xlsx',sheet_name=None,skiprows=[0,1,3])
df
```

```python
df_pop,df_spot=df.values()
```

```python
df_pop.head()
```

```python
df_day=df_pop[['조사지점코드','조사요일','유동인구수']]
df_day
```

```python
df_day=df_day.set_index('조사지점코드').join(df_spot.set_index('조사지점코드'))
df_day
```

```python
df_day=df_day.reset_index(drop=True).fillna('unknown')
df_day
```

```python
def createSeperator(sr):
    평일=('월','화','수','목','금')
    주말=('토','일')
    
    if sr['조사요일'] in 평일:
        return '평일'
    elif sr['조사요일'] in 주말:
        return '주말'
    else:
        return 'unknown'

df_day['평일/주말']=df_day.apply(createSeperator,axis=1)
df_day
```

```python
tmp=df_day.groupby(['조사지점명','평일/주말']).sum()
tmp
```

```python
df_weekEnd=tmp.xs('주말',level='평일/주말')
df_weekEnd
```

```python
df_weekDay=tmp.xs('평일',level='평일/주말')
df_weekDay
```

```python
df_weekDiff=abs(df_weekDay-df_weekEnd)
df_weekDiff
```

```python
# unknown 포함 출력
df_result=df_weekDiff.sort_values(by='유동인구수',ascending=False)
df_result[:5]
```

```python
# unknown 제외 출력
df_result=df_result.drop(['unknown'])
df_result[:5]
```

```python
df_result[:5].plot(kind = 'barh', fontsize = 13,figsize=(10, 10),colormap='Paired')
```
