```python
# X좌표, Y좌표를 위도와 경도로 변환하는 함수 선언
def convert(x,y):
    inProj = Proj(init = 'epsg:5181')
    outProj = Proj(init = 'epsg:4326')
    x2,y2 = transform(inProj, outProj, x, y)
    return [y2, x2]
```

```python
#위도와 경도 정보를 추가
df_tmpDong09['위경도'] = df_tmpDong09.apply(lambda 지점: convert(지점['X좌표'], 지점['Y좌표']), axis = 1)
df_tmpDong09.head(3)
```
