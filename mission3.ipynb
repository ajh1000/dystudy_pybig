```python
df_popTime=df_pop
```

```python
df_popTime=df_popTime.set_index('조사지점코드').join(df_spot.set_index('조사지점코드'))
df_popTime
```

```python
df_popTime=df_popTime.reset_index(drop=True)
```

```python
df_result=df_popTime[['조사지점명','유동인구수','시간대']]
df_result
```

```python
df_result=df_result.groupby(['조사지점명','시간대']).sum()
```

```python
#06시-10시
#18시-21시

df_result=df_result.reset_index().set_index('조사지점명')
df_result
```

```python
df_result['시간대']=df_result['시간대'].apply(lambda t : datetime.strptime( t.split('시-')[1],'%H시' ) )
df_result
```

```python
def createCommute(sr):
    if sr['시간대'] >=datetime(1900,1,1,7, 0, 0) and sr['시간대'] <= datetime(1900,1,1,10,0,0):
        return '출근'
    elif sr['시간대'] >=datetime(1900,1,1,19, 0, 0) and sr['시간대'] <=datetime(1900,1,1,21,0,0):
        return '퇴근'

df_result['출퇴근'] = df_result.apply(createCommute, axis=1)
df_result
```

```python
df_result=df_result[(df_result['출퇴근']=='출근') | (df_result['출퇴근']=='퇴근')]
df_result
```

```python
def resumeFormat(sr):
    beforeOneHour= sr['시간대']-timedelta(hours=1)
    
    strLeft=beforeOneHour.strftime('%H')
    strLeft+="시"
    
    strRight=sr['시간대'].strftime('%H')
    strRight+="시"
    
    return '-'.join([strLeft,strRight])

#passes a Series object, row-wise
df_result['시간대']=df_result.apply(resumeFormat,axis=1)
df_result
```

```python
df_result=df_result.reset_index().set_index(['조사지점명','출퇴근','시간대'])
```
