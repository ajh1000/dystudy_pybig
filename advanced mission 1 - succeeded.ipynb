```python
df_resultGu09=df_tmp09.reset_index().set_index('구코드').join(df_geoGu09[['구코드','구명']].set_index('구코드'))
df_resultGu09
```

```python
m09 = folium.Map(
    location=[37.566345, 126.977893],
    #tiles='cartodbpositron',
    zoom_start=11
)

rfile = open('data/skorea_municipalities_geo_simple.json', 'r', encoding='utf-8').read()
jsonData = json.loads(rfile)

m09.choropleth(
    geo_data=jsonData,
    name='choropleth',
    data=df_resultGu09,
    columns=['구명', '유동인구수'],
    key_on='feature.properties.name',
    fill_color='BuPu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='a floating population 2009'
)

folium.LayerControl().add_to(m09)

m09
```
